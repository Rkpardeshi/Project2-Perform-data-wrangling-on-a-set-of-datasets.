```python
import pandas as pd
```

```python
dataset3 = pd.read_csv("D:\\digicrome_datascienceAndAI\\Second_Project\\dataset_3 - dataset_3.csv")
new_data = pd.read_csv("D:\\digicrome_datascienceAndAI\\Second_Project\\new_data - new_data.csv")
```

```python
Final_data = pd.concat([new_data, dataset3], ignore_index=True)
```

```python
Final_data.loc[:, 'dteday'] = pd.to_datetime(Final_data['dteday'], dayfirst=True)
```

```python
print(Final_data)
```

```python
Final_data.head()
```

```python
Final_data = Final_data.fillna(Final_data.mean(numeric_only=True))
```

```python
# Step 1: Select only numeric columns
numeric_data = Final_data.select_dtypes(include='number')

# Step 2: Calculate IQR
Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3 - Q1

# Step 3: Filter out the outliers from the original DataFrame using numeric data
Final_data = Final_data[~((numeric_data < (Q1 - 1.5 * IQR)) | 
                          (numeric_data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print(Final_data.skew(numeric_only=True))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 12))
sns.heatmap(Final_data.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
